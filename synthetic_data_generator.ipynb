```python
import numpy as np
import pandas as pd
import random
import scipy.stats
import seaborn as sns
import matplotlib.pyplot as plt

def x(): 
    return np.random.normal(5, 3)

def f(x): 
    return x ** 2 + 1 + np.random.normal()

def y(df): 
    df['y'] = 2*df['x1'] + 3*df['x2'] - 5*df['x3'] + random.randrange(-5, 5)

N = 500

data = pd.DataFrame({'x1': [x() for i in range(N)]})
data['x2'] = f(data['x1'])
data['x3'] = [x() for i in range(N)]
data['x4'] = [x() for i in range(N)]
y(data)
# data

sns.pairplot(data)
plt.show()

sns.heatmap(data.corr(), annot=True)
plt.show()

data.to_csv("files/multi_reg.csv", index=False)
```
